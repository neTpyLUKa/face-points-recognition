```python
from keras.layers.advanced_activations import LeakyReLU
from keras.models import Sequential, Model
from keras.applications import MobileNetV2
from keras.layers import Activation, Convolution2D, MaxPooling2D, BatchNormalization, Flatten, Dense, Dropout, Conv2D,MaxPool2D, ZeroPadding2D, GlobalAveragePooling2D
from keras.layers import Conv2D,Dropout,Dense,Flatten
from keras.models import Sequential

def classify(model, test_img_dir):
    train_data = {}
    #coef = {}
    for img_path in os.listdir(test_img_dir):
        img = io.imread(test_img_dir + '/' + img_path)
        #coef[img_path] = img.shape[0] / 100
        img = transform.resize(img, (96,96,3))
        train_data[img_path]= img
    image_list = []
    #ind_coef = []
    names = sorted(train_data.keys())
    for img_path in names:
        image_list.append(train_data[img_path])
     #   ind_coef.append(coef[img_path])
    image_list = np.array(image_list, dtype = 'float')
    #ind_coef = np.array(ind_coef, dtype='float')
    X_test = image_list.reshape(-1,96,96,3)
    y_test = model.predict(X_test)
    ans = {}
    for i in range(len(y_test)):
        cur = 0.0
        ind=0
        for j in range(50):
            if y_test[i][j]>cur:
                cur=y_test[i][j]
                ind=j
        ans[names[i]] = ind
    return ans
```

```python
def train_detector(train_gt, train_img_dir, fast_train=True):
    epochs=50
    batch_size=100
    if(fast_train):
        epochs=1
        batch_size=1

    train_data = {}
 #   coef = {}
    sz=0
    for img_path in os.listdir(train_img_dir):
        sz += 1
        img = io.imread(train_img_dir + '/' + img_path)
        img = img.reshape((96,96, 3))
      #  coef[img_path] = img.shape[0] / 100
        train_data[img_path] = img
    
    train_points = train_gt
    
    sorted_data = []
    #ind_coef = []
    for img_path in sorted(train_data.keys()):
        sorted_data.append(train_data[img_path])
     #   ind_coef.append(coef[img_path])
    image_list = np.array(sorted_data,dtype = 'float')
    X_train = image_list.reshape(-1,96,96,3)
    
   # ind_coef = np.array(ind_coef, dtype='float')
    
    training = train_points.drop('filename',axis = 1)
    y_train = []
    for i in range(0,sz):
        y = training.iloc[i][0]
        l = [0 for i in range(50)]
        l[y]=1
        y_train.append(y)
    y_train = np.array(y_train,dtype = 'float')
    
    from keras.layers import Conv2D,Dropout,Dense,Flatten
    from keras.models import Sequential
    base_model = MobileNetV2(include_top=False, weights='imagenet', input_shape=(96,96,3))
    num_classes=50
    # create a custom top classifier
    x = base_model.output
    x = Dropout(0.4)(x)
    x = GlobalMaxPooling2D()(x)
    x = Dense(1280, activation='relu')(x)
    predictions = Dense(num_classes, activation='softmax')(x)
    model = Model(inputs=base_model.inputs, outputs=predictions)
    
    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    model.fit(X_train,y_train,epochs = epochs,batch_size = batch_size,validation_split = 0.2)

  
```
